```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define population data and sample size

```python
# Define the population data (e.g., heights in inches)
population_data = np.array([68, 70, 72, 66, 74, 71, 69, 75, 73, 67, 70, 72, 68, 70, 71, 69, 72, 74, 68, 70])

# Define the sample size
sample_size = 5
```

## Generate a random sample from the population

```python
# Generate a random sample from the population
sample = np.random.choice(population_data, size=sample_size, replace=False)
```

## Plot both population and sample data

```python
# Create a scatter plot for both population and sample data on the same plot
plt.figure(figsize=(8, 4))
plt.scatter(np.arange(len(population_data)), population_data, label='Population', marker='o', alpha=0.7)
plt.scatter(np.arange(len(sample)), sample, label='Sample', marker='x', color='green', alpha=0.7)

plt.title('Population vs Sample Data')
plt.xlabel('Index')
plt.ylabel('Height (inches)')
plt.legend()
plt.grid(True)

plt.show()
```

## Plot the population data

```python
# Plot the population data
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.hist(population_data, bins=10, edgecolor='black', alpha=0.7)
plt.title('Population Data')
plt.xlabel('Height (inches)')
plt.ylabel('Frequency')
```

## Plot the random sample

```python
# Plot the random sample
plt.subplot(1, 2, 2)
plt.hist(sample, bins=5, edgecolor='black', alpha=0.7, color='green')
plt.title('Random Sample')
plt.xlabel('Height (inches)')
plt.ylabel('Frequency')
```

## Display population and sample statistics

```python
def print_statistics(data, label):
    mean = np.mean(data)
    std_dev = np.std(data)
    variance = np.var(data)
    
    print(f"{label} Mean: {mean}")
    print(f"{label} Standard Deviation: {std_dev}")
    print(f"{label} Variance: {variance}\n")

# Print statistics for population and sample
print_statistics(population_data, "Population")
print_statistics(sample, "Sample")
```
